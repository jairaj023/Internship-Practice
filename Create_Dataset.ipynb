```python
import pandas as pd
```

```python
import random
```

```python
Departments = ["HR", "IT", "Sales", "Finance", "Marketing", "Operations"]
Gender = ["Male", "Female"]
Attrition = ["Yes", "No"]
Performance = ["Excellent", "High", "Average", "Low"]
Eduaction = ["High School", "Diploma", "Bachalor's", "Master's", "PhD"]
```

```python
data = []
```

```python
for i in range(1, 2001):
  data.append ({
      "Employee_ID": f"EMP{i:04}",
      "Department": random.choice(Departments),
      "Salary": random.randint(25000, 1200000),
      "Experience": random.randint(0, 25),
      "Gender": random.choice(Gender),
      "Attrition": random.choice(Attrition),
      "Performance": random.choice(Performance),
      "Education": random.choice(Eduaction)
  })
```

```python
df = pd.DataFrame(data)
```

```python
print(df.head())
```

```python
df.to_csv("Dataset.csv", index=False)
```
