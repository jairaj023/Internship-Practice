**KPI's**

```python
# Total Matches
total_matches = ipl.shape[0]
print("Total Matches : ", total_matches)
```

```python
# Highest Run Scorer
highest_runs = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)

print(highest_runs.head(10))

plt.figure(figsize=(10, 5))
highest_runs.head(10).plot(kind='bar', color='green')
plt.title("Highest Top 10 Run Scorers")
plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.xticks(rotation=45)
plt.show()
```

```python
# Highest Wicket Taker

wicket = deliveries[deliveries['player_dismissed'].notna()&(deliveries['dismissal_kind'] != 'run out')]
highest_wickets = wicket.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False)

print(highest_wickets.head(10))

plt.figure(figsize=(10, 5))
highest_wickets.head(10).plot(kind='bar', color='orange')
plt.title("Highest Top 10 wicket Takers in IPL")
plt.xlabel("Bowler")
plt.ylabel("Wickets")
plt.xticks(rotation=45)
plt.show()
```

```python
# Team Wins
team_wins = ipl['winner'].value_counts()

plt.figure(figsize=(10, 6))
team_wins.plot(kind='bar')
plt.title("Team Wins")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.xticks(rotation=90)
plt.show()
```

```python
# Venue Analysis
Venue = ipl['venue'].value_counts().head(10)

plt.figure(figsize=(10, 6))
Venue.plot(kind='bar')
plt.title("Top 10 Venue by Matches")
plt.xlabel("Venue")
plt.ylabel("Matches")
plt.xticks(rotation=90)
plt.show()
```

```python
# Toss Impact
toss = ipl.groupby('toss_winner')['winner'].count().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
toss.plot(kind='bar')
plt.title("Matches Won by Toss Winning Teams")
plt.xlabel("Teams")
plt.ylabel("Matches")
plt.xticks(rotation=90)
plt.show()
```

**New KPI's**

```python
# Missing Values Analysis

print("Missing Values in Match:")
print(ipl.isnull().sum())

print("Missing Values in Deliveries:")
print(deliveries.isnull().sum())
```

```python
# Duplicate Records

print("\nDuplicated Records in Matches:", ipl.duplicated().sum())

print("Duplicated Records in Deliveries:", deliveries.duplicated().sum())
```

```python
# Orange Cap Analysis

season_runs = deliveries.merge(ipl[["id", "season"]], left_on = "match_id", right_on = "id")

orange = season_runs.groupby(["season", "batsman"])["batsman_runs"].sum().reset_index()

orange_cap = orange.loc [orange.groupby("season")["batsman_runs"].idxmax()]

print("Orange Cap Winner : ")
print(orange_cap)
```

```python
# Purple Cap Analysis

season_wickets = wicket.merge(ipl[["id", "season"]], left_on = "match_id", right_on = "id")

Purple = season_wickets.groupby(["season", "bowler"])["player_dismissed"].count().reset_index()

Purple_cap = Purple.loc [Purple.groupby("season")["player_dismissed"].idxmax()]

print("Purple Cap Winner : ")
print(Purple_cap)
```

```python
# Team Wise Winning wins percentage

wins = ipl["winner"].value_counts()

played = pd.concat([ipl["team1"], ipl["team2"]]).value_counts()

wins_percentages = (wins / played * 100).sort_values(ascending = False)

plt.figure(figsize=(12, 6))
sns.barplot(x=wins_percentages.index, y=wins_percentages.values)
plt.title("Team Wise Winning percentage")
plt.ylabel("Winning Percetange(%)")
plt.xlabel("Teams")
plt.xticks(rotation=90)
plt.show()
```

```python
# Toss Winner vs Match Winner

ipl["Toss_Match_Same"] = ipl["toss_winner"] == ipl["winner"]

print(ipl["Toss_Match_Same"].value_counts())

sns.countplot(x="Toss_Match_Same", data=ipl)
plt.title("Toss Winner vs Match Winner")
plt.show()
```

```python
# Season Wise Champion

champion = ipl.groupby("season").last()[["winner"]]
print("Season Wise Champion : ")
print(champion)
```

```python
# Player of the Match

pom = ipl["player_of_match"].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=pom.index, y=pom.values)
plt.title("Player of the Match")
plt.xlabel("Awards")
plt.ylabel("Players")
plt.xticks(rotation=90)
plt.show()
```
